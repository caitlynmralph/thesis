```python
import requests
import pandas as pd

apiKey = "mfmECr57kCksL2MC"
```

```python
# get artist ids

artists = ["Arctic Monkeys",
"The 1975",
"Florence + The Machine",
"YUNGBLUD",
"Foals",
"Mura Masa",
"The Kooks",
"Catfish and the Bottlemen",
"Sam Fender",
"The Wombats",
"slowthai",
"Bombay Bicycle Club",
"Declan McKenna",
"Blossoms",
"Circa Waves",
"Easy Life",
"La Roux",
"IDLES",
"Sea Girls",
"Jade Bird",
"Fontaines D.C.",
"Frank Carter & The Rattlesnakes",
"Inhaler",
"The Magic Gang",
"IDER",
"RAT BOY",
"Sports Team",
"The Sherlocks",
"Blaenavon",
"Swim Deep",
"Sophie and the Giants",
"Drenge",
"Shopping",
"Otherkin",
"The Ninth Wave",
"Declan Welsh and The Decadent West",
"Palma Violets",
"Sfven",
"Lauran Hibberd",
"King Nun"]

ids = {}

for artist in artists:
    
    try:

        p = {'apikey': apiKey, 'query': artist}

        r = requests.get('https://api.songkick.com/api/3.0/search/artists.json', params=p)
        artistID = r.json()['resultsPage']['results']['artist'][0]['id']

        ids[artist] = artistID
        
    except:
        
        ids[artist] = "Not Found"
    
df = pd.DataFrame.from_dict(ids, orient="index")
df.to_csv("ids-test.csv")
```

```python
# get artist touring

artistIDs = ["66370",
"8969499",
"8054163",
"349047",
"3347901",
"6907314",
"8914304",
"171022",
"2360099",
"1352869",
"9133189",
"8484598",
"9231359",
"8494363",
"484476",
"4045756",
"585122",
"9418339",
"5335538",
"9125544"]

tourDates = {}

for artistID in artistIDs:
    
    page = 1
    running = True
    
    while running == True:
        
        try:
            
            p = {'apikey': apiKey, 'page': page}

            r = requests.get('https://api.songkick.com/api/3.0/artists/%s/gigography.json?' % artistID, params=p)

            dates = r.json()['resultsPage']['results']['event']

            pageDates = pd.DataFrame.from_records(dates)

            if page == 1:
                artistDates = pageDates
            else:
                artistDates = pd.concat([artistDates, pageDates])

            page=page+1
            
        except:
        
            running = False

    tourDates[artistID] = artistDates
    
tourDatesDF = pd.concat(tourDates)

tourDatesDF.to_csv("tour-dates-test.csv")
    
    
```
